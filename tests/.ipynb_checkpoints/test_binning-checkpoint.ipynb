```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score


def tree_binning_bootstrap_auc(
    df_train,
    df_val,
    feature,
    target,
    max_leaf_nodes_max=7,
    n_bootstrap=30,
    random_state=42
):
    np.random.seed(random_state)

    auc_results = {k: [] for k in range(2, max_leaf_nodes_max + 1)}
    splits_dict = {k: [] for k in range(2, max_leaf_nodes_max + 1)}

    # Drop nulls in validation set and align target
    X_val = df_val[[feature]].dropna()
    y_val = df_val.loc[X_val.index, target]

    for max_leaf in range(2, max_leaf_nodes_max + 1):
        for i in range(n_bootstrap):

            # Bootstrap sample
            df_boot = df_train.sample(
                frac=1,
                replace=True,
                random_state=random_state + i
            )

            # Drop nulls in feature and align target
            df_boot = df_boot[[feature, target]].dropna()
            X_boot = df_boot[[feature]]
            y_boot = df_boot[[target]]

            # Fit tree
            tree = DecisionTreeClassifier(
                max_leaf_nodes=max_leaf,
                min_samples_leaf=0.10,
                random_state=random_state + i
            )

            tree.fit(X_boot, y_boot)

            # Predict on validation
            y_pred_val = tree.predict_proba(X_val)[:, 1]
            auc = roc_auc_score(y_val, y_pred_val)
            auc_results[max_leaf].append(auc)

            # Store split thresholds
            thresholds = np.sort(
                tree.tree_.threshold[tree.tree_.threshold > -2]
            )
            splits_dict[max_leaf].append(thresholds)

    # Plot
    plt.figure(figsize=(10, 6))
    data = [auc_results[k] for k in sorted(auc_results)]
    plt.boxplot(data, positions=sorted(auc_results))
    plt.xlabel("max_leaf_nodes")
    plt.ylabel("ROC AUC (validation)")
    plt.title(f"Bootstrap ROC AUC by max_leaf_nodes for {feature}")
    plt.grid(False)
    plt.show()

    return_dict = {}
    return_dict["auc_results"] = auc_results
    return_dict["splits_dict"] = splits_dict

    return return_dict
```

```python
# Testa
from sklearn.datasets import load_breast_cancer
import pandas as pd

data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

display(df)
```

```python
# Separa em treino e teste
from sklearn.model_selection import train_test_split

df_train, df_val = train_test_split(
    df,
    test_size=0.3,
    random_state=42,
    stratify=df["target"]  # mantém proporção da classe
)

display(df_train)
```

```python
dict_mean_radius_bin_size = tree_binning_bootstrap_auc(
    df_train,
    df_val,
    feature = 'mean radius',
    target = 'target',
    max_leaf_nodes_max=7,
    n_bootstrap=30,
    random_state=42
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from typing import Union
from sklearn.tree import DecisionTreeClassifier


def get_tree_binning_summary(
    df: pd.DataFrame,
    feature: str,
    target: str,
    max_leaf_nodes: int,
    min_samples_leaf: Union[int, float] = 0.10,
    random_state: int = 1,
    plot: bool = True
):
    """
    Performs supervised univariate binning using a Decision Tree
    and optionally plots the mean target per bin.

    Returns
    -------
    dict with:
        - thresholds
        - bin_summary (DataFrame)
    """

    # -----------------------------
    # Basic validation
    # -----------------------------
    df_clean = df[[feature, target]].dropna()

    if df_clean[target].nunique() != 2:
        raise ValueError("Target must be binary.")

    if not np.issubdtype(df_clean[feature].dtype, np.number):
        raise ValueError("Feature must be numeric.")

    X = df_clean[[feature]]
    y = df_clean[target]

    # -----------------------------
    # Fit tree
    # -----------------------------
    tree = DecisionTreeClassifier(
        max_leaf_nodes=max_leaf_nodes,
        min_samples_leaf=min_samples_leaf,
        random_state=random_state
    )

    tree.fit(X, y)

    thresholds = np.sort(
        tree.tree_.threshold[tree.tree_.threshold > -2]
    )

    # -----------------------------
    # Create bins
    # -----------------------------
    bins = [-np.inf] + list(thresholds) + [np.inf]
    df_clean["bin"] = pd.cut(df_clean[feature], bins=bins)

    # -----------------------------
    # Aggregate statistics
    # -----------------------------
    summary = (
        df_clean
        .groupby("bin")[target]
        .agg(["count", "mean"])
        .rename(columns={"mean": "target_rate"})
        .reset_index()
    )

    # -----------------------------
    # Plot
    # -----------------------------
    if plot:
        plt.figure(figsize=(10, 5))
        plt.bar(summary["bin"].astype(str), summary["target_rate"])
        plt.xticks(rotation=45)
        plt.ylabel("Mean Target")
        plt.xlabel("Bins")
        plt.title(f"Mean Target by Tree Bins - {feature}")
        plt.tight_layout()
        plt.show()

    return {
        "thresholds": thresholds,
        "bin_summary": summary
    }
```
